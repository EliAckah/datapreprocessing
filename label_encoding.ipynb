```python
#import dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#loading the data
bcancer_data = pd.read_csv('/content/data.csv')
```

```python
#print the first 5 rows
bcancer_data.head()
```

```python
#find the count for our data labels
bcancer_data['diagnosis'].value_counts()
```

```python
#load the LabelEncoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(bcancer_data.diagnosis)
```

```python
#append our transformed encoded labels to the dataframe
bcancer_data['target'] = labels
```

```python
bcancer_data.head()
```

0 --> Benign

1 --> Malignant

```python
bcancer_data['target'].value_counts()
```

label encoding of iris data set

```python
#loading the data
iris_data = pd.read_csv('/content/iris_data.csv')
```

```python
#print the 1st 5 rows
iris_data.head()
```

```python
#find the counts for our data labels
iris_data['Species'].value_counts()
```

```python
#load the LabelEncoder function
iris_encode = LabelEncoder()
```

```python
iris_labels = iris_encode.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_labels
```
